```python
import pandas as pd
import numpy as np
import os
stage1_folder = "stage1_organ_segmentator_eval"
pd.set_option("display.max_rows", None)
```

```python
preds = os.listdir(stage1_folder)
```

```python
def display_preds(preds_dataset_name, threshold_recall=0.9, threshold_iou=0.9):
    preds_folder = os.path.join(stage1_folder, preds_dataset_name)
    preds_metrics = pd.read_csv(os.path.join(preds_folder, "entrywise_preds.csv"), index_col=0)
    bad_slices = ((preds_metrics["slice_recall_liver"] < threshold_recall) & (preds_metrics["slice_iou_liver"] < threshold_iou))\
                    | ((preds_metrics["slice_recall_spleen"] < threshold_recall) & (preds_metrics["slice_iou_spleen"] < threshold_iou))\
                    | ((preds_metrics["slice_recall_kidney"] < threshold_recall) & (preds_metrics["slice_iou_kidney"] < threshold_iou))
    bad_preds = preds_metrics.loc[bad_slices]
    bad_preds = bad_preds[[col for col in bad_preds.columns if "recall" in col]]
    print(len(bad_preds))
    display(bad_preds)

def display_preds_highiou(preds_dataset_name, threshold_recall=0.9):
    preds_folder = os.path.join(stage1_folder, preds_dataset_name)
    preds_metrics = pd.read_csv(os.path.join(preds_folder, "entrywise_preds.csv"), index_col=0)
    bad_slices = ((preds_metrics["slice_recall_liver"] < threshold_recall) & (preds_metrics["slice_iou_liver"] == 1.0))\
                    | ((preds_metrics["slice_recall_spleen"] < threshold_recall) & (preds_metrics["slice_iou_spleen"] == 1.0))\
                    | ((preds_metrics["slice_recall_kidney"] < threshold_recall) & (preds_metrics["slice_iou_kidney"] == 1.0))
    bad_preds = preds_metrics.loc[bad_slices]
    bad_preds = bad_preds[[col for col in bad_preds.columns if ("recall" in col and "slice" in col) or "iou" in col]]
    print(len(bad_preds))
    display(bad_preds)

def display_preds_onlyhighiou(preds_dataset_name):
    preds_folder = os.path.join(stage1_folder, preds_dataset_name)
    preds_metrics = pd.read_csv(os.path.join(preds_folder, "entrywise_preds.csv"), index_col=0)
    bad_slices = ((preds_metrics["slice_iou_liver"] == 1.0))\
                    | ((preds_metrics["slice_iou_spleen"] == 1.0))\
                    | ((preds_metrics["slice_iou_kidney"] == 1.0))
    bad_preds = preds_metrics.loc[bad_slices]
    bad_preds = bad_preds[[col for col in bad_preds.columns if ("recall" in col and "slice" in col) or "iou" in col]]
    print(len(bad_preds))
    display(bad_preds)
```

# Mixed outliers

```python
for preds_dataset_name in preds:
    print(preds_dataset_name)
    display_preds(preds_dataset_name)
```

```python
for preds_dataset_name in preds:
    print(preds_dataset_name)
    display_preds(preds_dataset_name, threshold_recall=0.8)
```

# IOU outliers small

```python
for preds_dataset_name in preds:
    print(preds_dataset_name)
    display_preds_highiou(preds_dataset_name, threshold_recall=0.8)
```

# Only IOU outliers

```python
for preds_dataset_name in preds:
    print(preds_dataset_name)
    display_preds_onlyhighiou(preds_dataset_name)
```
