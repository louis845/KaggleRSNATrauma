```python
import pandas as pd
from IPython.display import display, Markdown
```

```python
def display_shapes(shapes):
    h_first = shapes.sort_values("shape_h", ascending=False)
    w_first = shapes.sort_values("shape_w", ascending=False)
    area = shapes.sort_values("area", ascending=False)

    display(Markdown("## H"))
    display(h_first.head(30))
    display(Markdown("## W"))
    display(w_first.head(30))
    display(Markdown("## Area"))
    display(area.head(30))
```

# Shapes

```python
shapes = pd.read_csv("data_npy/shape_info.csv")
shapes["area"] = shapes["shape_h"] * shapes["shape_w"]

display_shapes(shapes)
```

# Cropped shapes

```python
shapes = pd.read_csv("reduced_shape.csv")

display_shapes(shapes)
```
