```python
import pandas as pd
```

```python
shapes = pd.read_csv("data_npy/shape_info.csv")
```

```python
shapes["area"] = shapes["shape_h"] * shapes["shape_w"]
```

```python
h_first = shapes.sort_values("shape_h", ascending=False)
display(h_first)
```

```python
w_first = shapes.sort_values("shape_w", ascending=False)
display(w_first)
```

```python
area = shapes.sort_values("area", ascending=False)
display(area)
```
